```python
# test setup with captured data
test_payload = ['20  |Z|=10000.00\r\n',
                '30  |Z|=5883.48\r\n',
                '40  |Z|=10000.00\r\n',
                '50  |Z|=18027.76\r\n',
                '   \r\n',
                '20  |Z|=10000.00\r\n',
                '30  |Z|=9486.83\r\n',
                '40  |Z|=8687.45\r\n',
                '50  |Z|=18027.76\r\n',
                '   \r\n',
                '20  |Z|=10000.00\r\n',
                '30  |Z|=7276.07\r\n',
                '40  |Z|=10000.00\r\n',
                '50  |Z|=18027.76\r\n']

# libraries
import serial
import time
import pandas as pd
import matplotlib
import datetime

# variables to set
# capture time in seconds
capture_duration = 2            

# setup
scan = 0
payload = []
def fake_read():
    global scan
    time.sleep(0.05)
    data = test_payload[scan%len(test_payload)]
    scan += 1
    return data

t_end = time.time() + capture_duration
while time.time() < t_end:
    value = fake_read()
    payload.append(value) # printing the value
    
#clean and format
df = pd.DataFrame(payload)
df.columns = ['value']
df['group'] = df['value'].str.split("  ", expand=True)[0]
df['impedence'] = pd.to_numeric(df['value'].str.split("  ", expand=True)[1].str.rstrip('\r\n').str.split('=', expand = True)[1])
filtered_data = df[['group','impedence']].dropna()

#export
!mkdir captures
csv_path = 'captures/test_capture_'+datetime.datetime.now().strftime("%d-%m-%Y_%H:%M:%S")+'.csv'
filtered_data.to_csv(csv_path, index=False)

#plot
filtered_data.groupby('group').plot()
```
